```python
from pylab import figure, cm

import matplotlib.pyplot as plt
import numpy as np

t_min = 0
t_max = 5
x_min = 0
x_max = 5

data = np.loadtxt('sol.txt', ndmin=2)

plt.imshow(data,extent=[x_min,x_max,t_min,t_max], cmap=cm.jet, origin='lower')
plt.colorbar()
plt.title("Transfer equation solution" , fontsize=14)
plt.xlabel("x")
plt.ylabel("t")
plt.savefig("sol.png", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

times = np.array([31.91, 21.24, 13.62, 12.68])
speedup = times[0] / times 
num_proc = np.linspace(1.0, 4.0, num=4)

plt.plot(num_proc, speedup)
plt.title("MPI Speedup" , fontsize=14)
plt.xlabel("Number of processes")
plt.ylabel("Speedup")
plt.show()
```
