```python
import numpy as np
from statistics import mean 
import matplotlib.pyplot as plt

ref_time = mean([0.193025, 0.184984, 0.202441, 0.181883, 0.177835])

time = []
time.append(ref_time / mean([0.200553, 0.184478, 0.186013, 0.194133, 0.185536]))
time.append(ref_time / mean([0.121136, 0.112724, 0.117155, 0.118103, 0.125553]))
time.append(ref_time / mean([0.111068, 0.10769, 0.12322, 0.143324, 0.109275]))
time.append(ref_time / mean([0.0937078, 0.0926922, 0.105469, 0.0876812, 0.0888138]))

plt.plot(time)
plt.xticks(np.arange(len(time)), np.arange(1, len(time)+1))
plt.title('00 MPI speedup')
plt.xlabel('Number of processes')
plt.ylabel('Speedup')

plt.show()
```
