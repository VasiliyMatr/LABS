```python
from statistics import mean 
import matplotlib.pyplot as plt

ref_time = mean([1.16125, 1.18265, 1.2827, 1.3198, 1.31814])

time = []
time.append(ref_time / mean([1.17987, 1.47381, 1.5162, 1.37502, 1.36472]))
time.append(ref_time / mean([0.843998, 0.898623, 0.90565, 1.05407, 1.12131]))
time.append(ref_time / mean([0.753222, 0.64445, 0.893529, 1.1691, 0.964803]))
time.append(ref_time / mean([0.652817, 0.703945, 0.731193, 0.690056, 0.662814]))

plt.plot(time)
plt.title('03 speedup')
plt.xlabel('Number of processes')
plt.ylabel('Speedup')

plt.show()
```
