```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

C = 3e8
Me = 9.11e-31

K = 235 / C

I = np.arange(start=0.2, stop=4 + 0.01, step=0.2)
I = np.append(I, np.arange(4.1, 4.4 + 0.01, 0.1))
I = np.append(I, np.arange(4.6, 5.0 + 0.01, 0.2))

N = np.array([0.690, 0.720, 0.670, 0.660, 1.230, 2.039, 4.049, 6.828, 8.847, 11.387, 11.327, 11.727, 11.087, 9.837, 9.027, 7.388, 4.349, 2.499, 1.959, 3.169, 8.787, 14.316, 15.306, 10.317, 1.769, 0.560, 0.390])
N = N - 0.7498

P = K * I * 1e3 * 1.6e-19
E = np.sqrt(P**2 * C**2 + Me**2 * C**4) - Me * C**2
E = E / 1.6e-19 / 1e3

print(P)

f = np.sqrt(N) / (I*235 / 10000)**(3/2)

plt.figure(figsize=[12, 5], dpi=100)
plt.plot(I, N, 'b.')

plt.xlabel('\huge I, A')
plt.ylabel('\huge N, 1/c')
plt.grid()

plt.show()

plt.figure(figsize=[12, 5], dpi=100)
plt.plot(E, f, 'r.')

plt.xlabel('\huge E, keV')
plt.ylabel('\huge f')
plt.grid()

plt.show()
```
