```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

T = np.array([920, 1030, 1140, 1250, 1350, 1450, 1550, 1660, 1770, 1880])
I = np.array([0.684, 0.761, 0.851, 0.961, 1.056, 1.148, 1.227, 1.352, 1.450, 1.548])
V = np.array([22.3, 28.2, 35.7, 45.8, 54.7, 64.5, 73.1, 87.4, 99.4, 105.4])
W = I * V

lnW = np.log(W)
lnT = np.log(T)

[a, b] = np.polyfit(lnT, lnW, 1)

plt.figure(figsize=[12, 5], dpi=100)

plt.plot(lnT, lnW, 'b.')
plt.plot(lnT, a*lnT + b, '-')

plt.xlabel("$ ln(T) $")
plt.ylabel("$ ln(W) $")
plt.title("W(T)")
plt.grid()

plt.show()

print(W)
print(a, b)
```
