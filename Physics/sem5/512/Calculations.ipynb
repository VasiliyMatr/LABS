```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

Theta = np.arange(start=0, stop=121, step=10)
N = np.array([900, 856, 750, 716, 641, 581, 498, 454, 397, 370, 344, 300, 281])

x = 1 - np.cos(Theta * np.pi / 180)
NInverse = 1 / N
[a, b] = np.polyfit(x, NInverse, 1)

plt.figure(figsize=[12, 5], dpi=100)

plt.plot(x, NInverse, 'b.')
plt.plot(x, a*x + b, '-')

plt.xlabel("$ 1 - cos(\Theta) $")
plt.ylabel("$ 1/N $")
plt.title("$ 1/N ( 1 - cos(\Theta) ) $")
plt.grid()

exp_quad_error = 0 #np.square(np.subtract(a*x + b, NInverse)).mean()
mes_epsilon = 10 / 400

N_0 = 1 / b
N_90 = 1 / (b + a*1)
N_0_sigma = np.sqrt (exp_quad_error + (mes_epsilon * N_0) ** 2) 
N_90_sigma = np.sqrt (exp_quad_error + (mes_epsilon * N_90) ** 2) 

print ("coeffs: ", a, b)
print ("N(0): ", N_0)
print ("N(90):", N_90)
print ("error N(0): ", N_0_sigma)
print ("error N(90): ", N_90_sigma)

plt.show()
```
