```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

Al_N = np.array([219904, 139859, 89422, 58460, 38075])
Al_l = np.array([19.8, 39.75, 59.55, 79.5, 99.5]) * 0.1

Fe_N = np.array([183792, 99011, 53681, 30113, 16662])
Fe_l = np.array([10, 19.95, 29.95, 39.9, 49.9]) * 0.1

Pb_N = np.array([185379, 98657, 55228, 30396, 17977])
Pb_l = np.array([4.8, 9.7, 14.4, 19, 23.4]) * 0.1

N0 = 379077

Al_ln = np.log(N0 / Al_N)
Fe_ln = np.log(N0 / Fe_N)
Pb_ln = np.log(N0 / Pb_N)

Al_a, Al_b = np.polyfit(Al_l, Al_ln, 1)
Fe_a, Fe_b = np.polyfit(Fe_l, Fe_ln, 1)
Pb_a, Pb_b = np.polyfit(Pb_l, Pb_ln, 1)

plt.figure(figsize=[12, 5], dpi=100)
plt.plot(Al_l, Al_ln, 'b.')
plt.plot(Al_l, Al_a*Al_l+Al_b, 'r')

plt.xlabel('\huge l, cm')
plt.ylabel('\huge ln(N0/N)')
plt.grid()
plt.show()

plt.figure(figsize=[12, 5], dpi=100)
plt.plot(Fe_l, Fe_ln, 'b.')
plt.plot(Fe_l, Fe_a*Fe_l+Fe_b, 'r')

plt.xlabel('\huge l, cm')
plt.ylabel('\huge ln(N0/N)')
plt.grid()
plt.show()

plt.figure(figsize=[12, 5], dpi=100)
plt.plot(Pb_l, Pb_ln, 'b.')
plt.plot(Pb_l, Pb_a*Pb_l+Pb_b, 'r')

plt.xlabel('\huge l, cm')
plt.ylabel('\huge ln(N0/N)')
plt.grid()
plt.show()

print("Al_a: ", Al_a)
print("Fe_a: ", Fe_a)
print("Pb_a: ", Pb_a)
```

```python
#   :

import numpy as np

np.set_printoptions(precision=4)

Marks = [9,8, 9,7, 9,7, 9,8, 9,9, 8,8, 8,8, 9,8]
# Karina:                 ^=9

Avg = np.mean(Marks)
print("Average: {:.2f}".format(Avg))

Sigma = np.sqrt(1/(len(Marks)**2) * np.sum((Marks - Avg)**2))
print("Sigma: {:.2f}".format(Sigma))
print("Lower: {:.2f}".format(Avg - Sigma))
print("Upper: {:.2f}\n".format(Avg + Sigma))
print("Suggested: 9.00")
```
